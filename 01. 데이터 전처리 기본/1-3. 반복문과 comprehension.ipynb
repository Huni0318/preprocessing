#### itertools 모듈 사용

```python
import itertools
```

```python
# itertools.product
count = 0
for a, b, c in itertools.product(range(5),range(5),range(5)):
    print(a,b,c)
```

```python
for i,j in itertools.product(range(2,10),range(2,10)):
    print("{}*{}={}".format(i, j, i*j))
```

```python
# 위 코드와 같은 방식으로 동작하는 코드
for a in range(5):
    for b in range(5):
        for c in range(5):
            print(a, b, c)
```

```python
# zip
for a, b, c in zip(range(5), range(5), range(5)):
    print(a, b, c)
```

```python
# itertools.combination
# L에서 2개로 이뤄진 가능한 모든 조합 뽑기
L =['a','b','c','d']
for comb in itertools.combinations(L,2):
    print(comb)
```

```python
# L에서 가능한 모든 조합 뽑기
for r in range(1, len(L)+1):
    for comb in itertools.combinations(L,r):
        print(comb)
```

```python
# L에서 가능한 모든 순열 뽑기
for r in range(1, len(L)):
    for comb in itertools.permutations(L, r):
        print(comb)
```

#### List comprehension

```python
L1 = [x for x in range(10) if x%2 == 0]
L1
```

```python
L1 = []
for x in range(10):
    if x%2 == 0:
        L1.append(x)
L1
```

```python
L2 = [x**2 for x in range(10)]
L2
```

```python
L3 = [x**2 for x in range(10) if x%2 == 0]
L3
```

```python
L3 = []
for x in range(10):
    if x%2 == 0:
        L3.append(x**2)
L3
```

```python
X = [1, 2, 3, 'a', 'b']
Y = [3, 1, 2, 'c', 4]

L4 = [x + y for x, y in itertools.product(X, Y) if type(x) == type(y) == int]
L4
```

#### dictionary comprehension

```python
dic1 = {x:y for x, y in zip(range(10), range(10))}
dic1
```

```python
dic = {1:'NaN', 2:2, 3:4, 4:'NaN'}
dic2 = {x:y for x, y in dic.items() if y != 'NaN'}
dic2
```

```python
dic2 = dict()
for x, y in dic.items():
    if y != 'NaN':
        dic2[x] = y
```
