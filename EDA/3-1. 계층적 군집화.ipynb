```python
df = pd.read_csv('Telco_customer_info.csv')
df.head()
```

```python
df.set_index('customerID', inplace = True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first = True)
df.head()
```

```python
df.columns
```

```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward').fit(df)
```

```python
df[''] = clusters.labels_
df[''].head()
```

```python
df.groupby([''])[['MonthlyCharges', 'TotalCharges']].mean()
```

```python
df.groupby([''])[['SeniorCitizen', 'StreamingTV_Yes']].mean()
```

```python
df = pd.read_csv('_.txt', sep = '\t', engine = 'python')
```

```python
df.head()
```

```python
matrix_df = pd.crosstab(index = df[''], columns = df[''])
matrix_df.head()
```

```python
#     
from sklearn.cluster import AgglomerativeClustering as AC
clustering_model = AC(n_clusters = 10,
                      affinity = "jaccard",
                      linkage = "average")
clustering_model.fit(matrix_df)
```

```python
cluster_labels = clustering_model.labels_
cluster_labels
```

```python
#    
cluster_info = pd.DataFrame({"ID":matrix_df.index, "":cluster_labels})
cluster_info.head()
```

```python
cluster_info[''].value_counts() #  0  
```

```python
matrix_df_with_cluster_info = pd.merge(matrix_df, cluster_info, left_index = True, right_on = 'ID')
matrix_df_with_cluster_info
```

```python
matrix_df_with_cluster_info.groupby([''])[matrix_df.columns].mean()
```

```python
#       
matrix_df_with_cluster_info.groupby([''])[matrix_df.columns].mean().idxmax(axis = 1)
```
